## Training and test split

The training dataset is going to be the the data from 2010-2011. The test data is going to be the following year, i.e, 2012

```python
dataset['Date']=pd.to_datetime(dataset['Date'], dayfirst = True)
dataset['year']=dataset['Date'].dt.year
dataset['month']=dataset['Date'].dt.month
dataset['day']=dataset['Date'].dt.day
```

```python
train = dataset.loc[dataset['year'] < 2012]
test = dataset.loc[dataset['year'] >= 2012]

X_train = train.drop(['Date','Weekly_Sales',"CPI", "Fuel_Price", 'Unemployment'],axis=1).values
y_train = train['Weekly_Sales'].values
X_test = test.drop(['Date','Weekly_Sales',"CPI", "Fuel_Price", 'Unemployment'],axis=1).values
y_test = test['Weekly_Sales'].values
```

## Random Forest Regression

```python
# Fitting Random Forest Regression to the dataset
from sklearn.ensemble import RandomForestRegressor
regressor = RandomForestRegressor(n_estimators = 100, random_state = 0)
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
print(r2_score(y_test, y_pred))
print(mean_absolute_error(y_test, y_pred))
print(np.sqrt(mean_squared_error(y_test, y_pred)))
print(np.std(test['Weekly_Sales']))
```

```python
plt.scatter(y_test,y_pred)
plt.title("Original data")
plt.xlabel("Real Value")
plt.ylabel("Predicted value")
plt.show()
```

Both the MAE and RMSE has low values and the R2 Score is closer to 1 indicating that it has performed well for our dataset.

Plotting y_test and y_pred to check comparison of forecast. We can see that Random Forest has performed pretty well in its forecasting.

```python
plt.plot(y_test)
plt.plot(y_pred, color = 'red')
```
